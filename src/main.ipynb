### Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('include/nyc-rolling-sales.csv', index_col=0, delimiter=',')
df.info()
df.head(15)
```

```python
# Bytter ut white space med NaN
df = df.replace(r'^\s*$', np.nan, regex=True)
df = df.replace(r'^-*$', np.nan, regex=True)

df.head(15)
```

```python
print("Sjekker om noen kolonner er tomme:" + "\n", df.isna().sum())
```

```python
# Fjerner irrelevante kolonner fra datasett
# Vi kan også endre verdi på Borough (1-5) til
# Manhattan (1), Bronx (2), Brooklyn (3), Queens (4), and Staten Island (5).

del df["EASE-MENT"]  #Tom, burde fjernes
del df["BUILDING CLASS AT PRESENT"]
del df["LOT"]
del df["SALE DATE"]
del df["TAX CLASS AT PRESENT"]
del df["ZIP CODE"]
del df["BLOCK"]
del df["COMMERCIAL UNITS"]
del df["TOTAL UNITS"]
df.head(15)
```

```python
#Fjerner rader hvor sale price er tom
df = df[df["SALE PRICE"].str.contains("-")==False]
#df = df[df["SALE PRICE"].str.contains("0")==False]
```
